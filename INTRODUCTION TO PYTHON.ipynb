***What is python?***

Python is a platform independent, object oriented scripting language, ready to run any type of program, from Windows applications to network servers or even web pages. It is an interpreted language, which means that it is not necessary to compile the source code to be able to execute it, which offers advantages such as speed of development and the reduction of errors when programming.

***Features***



*   You can create all kinds of programs, from the most basic to complex web pages.
*   There are versions of Python available on many different computer systems, allowing more flexibility in its use.
*   The code should not be compiled before running, which reduces the number of errors when programming.
*   Object-oriented programming is compatible with Python and offers an easy way to create programs.

****Python history****

Guido van Rossum devised the Python language in the late 1980s and began implementing it in December 1989. In February 1991, he published the first public version, version 0.9.0. Version 1.0 was released in January 1994, version 2.0 was released in October 2000, and version 3.0 was released in December 2008.

Until 2018, Python development was personally led by Guido van Rossum and under the umbrella of the Python Software Foundation. In July 2018, Guido van Rossum announced that he would stop leading Python development. Since 2019, Python development has been led by a five-member board of directors elected from among Python developers and will be renewed annually.

**Loops**

A loop is a sequence that repeats the same code snippet several times, until the condition assigned to the loop is no longer met.
Loops allow us to execute the same code several times, until the condition that causes the loop to execute is no longer met. And we have basically 3 types of loops, the while, the for and the do while.

>While loop example:

```python
i = 1
while i <= 3:
    print(i)
    i += 1
print("Programa terminado")
```

>For loop example:

```python
print("Comienzo")
for i in [0, 1, 2]:
    print("Hola ", end="")
print()
print("Final")
```

>Do while loop example:

```python
contador = 0
while True:
    print(contador)
    contador += 1
    if contador > 5:
        break
```
